```python
# PDF
import pdfplumber
import pandas as pd

def extrair_texto_e_tabela(DadosSinteticos):
    """Extrai texto e a tabela de um arquivo PDF."""

    with pdfplumber.open(DadosSinteticos) as pdf:
        primeira_pagina = pdf.pages[0]

        # Extrai todo o texto da primeira página
        texto_completo = primeira_pagina.extract_text()
        print("\n--- Texto Extraído ---")
        print(texto_completo)

        # Extrai a primeira tabela encontrada
        tabelas = primeira_pagina.extract_tables()

        if tabelas:
            print("\n--- Tabela Extraída ---")
            tabela_encontrada = tabelas[0]

            # Itera sobre as linhas da tabela para imprimir os dados
            for linha in tabela_encontrada:
                print(linha)
        else:
            print("\nNenhuma tabela foi encontrada no PDF.")

if __name__ == "__main__":
    extrair_texto_e_tabela("DadosSinteticos-personas1.pdf")

```

```python
#estatistica descritiva

arquivo = "DadosSinteticos-personas.xlsx"
df = pd.read_excel(arquivo)

# Estatística descritiva completa (numéricas + categóricas)
estatisticas = df.describe(include='all').transpose()

# Para as numéricas, adiciona Coeficiente de Variação
for col in df.select_dtypes(include='number').columns:
    estatisticas.loc[col, 'CV'] = estatisticas.loc[col, 'std'] / estatisticas.loc[col, 'mean']

print("\n--- Estatística Descritiva Completa ---")
print(estatisticas)

```

```python
import pandas as pd
import plotly.express as px

# Lê a planilha
arquivo = "DadosSinteticos-personas.xlsx"
df = pd.read_excel(arquivo)

# Dicionário de subcategorias
categorias_dores = {
    "Comunicação": ["comunicação", "informações dispersas", "aprovações demoradas", "linguagem técnica"],
    "Gestão": ["retrabalho", "falta de padronização", "prazos", "KPIs", "alertas automáticos", "centralização"],
    "Dados": ["dados ausentes", "valores duplicados", "erros de digitação", "formatação", "consolidar"],
    "Acesso/Bi": ["falta de acesso", "extração manual", "bases inconsistentes"]
}

# Função para classificar cada dor
def classificar_dores(texto):
    categorias = []
    texto_lower = str(texto).lower()
    for cat, palavras in categorias_dores.items():
        for palavra in palavras:
            if palavra in texto_lower:
                categorias.append(cat)
                break
    return categorias if categorias else ["Outros"]

# Criar coluna com categorias de dor
df["Subcategorias Dores"] = df["Dores"].apply(classificar_dores)

# Explode para ter uma linha por subcategoria
df_exp = df.explode("Subcategorias Dores")

# Contagem de dores por cargo e subcategoria
df_contagem = df_exp.groupby(["Cargo", "Subcategorias Dores"]).size().reset_index(name="Contagem")

# Gráfico de barras interativo
fig = px.bar(
    df_contagem,
    x="Subcategorias Dores",
    y="Contagem",
    color="Cargo",
    barmode="group",
    title="Distribuição de tipos de insatisfação por cargo",
    text="Contagem"
)
fig.show()
```
